```python
import pandas as pd

N_xlsx_url = "C:\\Users\\ymg08\\Desktop\\python_code\\food_menu_refer\\nutrient.xlsx"
kimchi_xlsx_url = "C:\\Users\ymg08\\Desktop\\python_code\\food_menu_refer\\kimchi.xlsx"
soup_xlsx_url = "C:\\Users\\ymg08\\Desktop\\python_code\\food_menu_refer\\soup.xlsx"
sub1_xlsx_url = "C:\\Users\\ymg08\\Desktop\\python_code\\food_menu_refer\\submenu1.xlsx"

N_df = pd.read_excel(N_xlsx_url, usecols=[1, 7, 9, 17, 19, 20, 22]) # 분류 추출
N_df.columns = ['menu1', 'category', 'vv', 'kcal', 'protein', 'lipid', 'carbohydrates']
kimchi_df = pd.read_excel(kimchi_xlsx_url, usecols=[1])
#kimchi_df
N_df
```

```python
N_raw_df = N_df[N_df['category'] == '김치류']
```

```python
N_dic = N_raw_df.to_dict('records')
N_dic
```

```python
kim_dic = kimchi_df.to_dict('records')
kim_dic
```
