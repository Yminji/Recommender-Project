```python
N_soup_df = N_df[(N_df['category'] == '볶음류') | (N_df['category'] == '조림류') | (N_df['category'] == '나물·숙채류') | (N_df['category'] == '생채·무침류') | (N_df['category'] == '장류, 양념류') | (N_df['category'] == '구이류') | (N_df['category'] == '찜류') | (N_df['category'] == '전·적 및 부침류') | (N_df['category'] == '튀김류') | (N_df['category'] == '면 및 만두류') | (N_df['category'] == '장아찌·절임류')]
```

```python
N_soup_df   
```

```python
N_dic = N_soup_df.to_dict('records')
N_dic
```

```python
sub1_dic = sub1_df.to_dict('records')
sub1_dic
```

```python
new_data = []

for item1 in N_dic:
    for item2 in sub1_dic:
        if item1.get('vv') in item2.get('menu'):
            new_data.append(item1 | item2) # | : 리스트 합치는 용도
            
print(new_data)
```

```python
import re

dd = []

for item in new_data:
    menu = str(item['menu1'])
    dd = re.sub(r'.*_', '',menu)
    item['menu1'] = dd
    
   # print(item['menu1'])

new_data
```
