```python
import pandas as pd

sub1_xlsx_url = "C:\\Users\\ymg08\\Desktop\\python_code\\food_menu_refer\\kimchi.xlsx"
N_xlsx_url = "C:\\Users\\ymg08\\Desktop\\python_code\\food_menu_refer\\nutrient.xlsx"


N_df = pd.read_excel(N_xlsx_url, usecols=[1,2,3,4,5,6,7])
N_df.columns = ['menu1', 'category', 'vv', 'kcal', 'protein', 'lipid', 'carbohydrates']
N_df.head()
```

```python
sub1_df = pd.read_excel(sub1_xlsx_url, usecols=[1,2,3])
sub1_df.head()
```

```python
N_soup_df = N_df[(N_df['category'] == '김치류') ]
```

```python
N_soup_df.head()
```

```python
N_dic = N_soup_df.to_dict('records')
N_dic[:3]
```

```python
sub1_dic = sub1_df.to_dict('records')
sub1_dic[:3]
```

```python
new_data = []

for item1 in N_dic:
    for item2 in sub1_dic:
        if item1.get('vv') in item2.get('menu'):
            new_data.append(item1 | item2) # | : 리스트 합치는 용도
            
print(new_data[:3])
```

```python
import re

dd = []

for item in new_data:
    menu = str(item['menu1'])
    dd = re.sub(r'.*_', '',menu)
    item['menu1'] = dd
    
   # print(item['menu1'])

new_data[:3]
```
