```python
import pandas as pd
xlsx_name = 'C:\\Users\\ymg08\\Desktop\\python_code\\식단 추천 알고리즘 프로젝트\\국_찌개.xlsx'
data = pd.read_excel(xlsx_name)

data.columns=['menu', 'material', 'amount']
data
```

```python
data.to_dict('records')
```

```python

# 'menu'를 키로 사용하여 결과를 저장할 딕셔너리 생성
result_dict = {}

# 데이터를 순회하면서 딕셔너리 생성
for _, row in data.iterrows():
    menu = row['menu']
    # 'materal'과 'amount'값을 문자열로 변환
    material = str(row['material']) 
    amount = str(row['amount'])
    
    # 'menu' 값이 NaN이 아닌 경우 새로운 딕셔너리를 생성하고 딕셔너리에 추가
    if not pd.isna(menu):
        result_dict[menu] = {'menu': menu, 'material': material, 'amount': amount}
    else:
        # NaN인 경우, 이전 'menu'의 값을 이용하여 'material'과 'amount' 정보 추가
        if result_dict:
            prev_menu = list(result_dict.keys())[-1]
            result_dict[prev_menu]['material'] += ', ' + material
            result_dict[prev_menu]['amount'] += ',' + amount

# 딕셔너리 값을 리스트로 변환하여 최종 결과 얻음
result_list = list(result_dict.values())

print(result_list)
```

```python
data_result = pd.DataFrame(result_list)

data_result
```

```python
data_result.to_excel('soup.xlsx')
```
